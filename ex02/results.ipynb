```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
files = ['part-00000', 'part-00001', 'part-00002']

list_df = []
for file in files:
    _df = pd.read_table(file, index_col=None, header=None)
    list_df.append(_df)
df = pd.concat(list_df, axis=0, ignore_index=True)
df.columns = ['year', 'max_temp']

df['max_temp'] = df['max_temp'].apply(lambda t: t/10)

df = df.sort_values('year')
print(df.shape)
df
```

```python
plt.figure(figsize=(12,8))

plt.plot(df['year'], df['max_temp'], 'x-', color='r')

plt.xticks(range(1973, 2015+1, 3))
plt.grid(True, which='both', axis='both')
plt.ylim(0, 40)

plt.title('Max temperature in Switzerland over the years')
plt.xlabel('Year')
plt.ylabel('Temperature [°C]')

plt.show()
```
